```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.arange(0,4*np.pi, 0.1)
y=np.sin(x)
```

```python
plt.figure(figsize=(8,4))

plt.plot(x,y,linewidth=2,linestyle="-",color='#8E44AD',marker="^")
plt.title("sin(x) function")
plt.xlabel("pi")
plt.grid(color="k",linestyle="--")
plt.show()
```

```python
x = np.arange(0, 4 * np.pi, 0.1)
sin = np.sin(x) # sin(x)
cos = np.cos(x) # cos(x)

plt.figure(figsize=(12,6))
# sin(x) grafigi
plt.plot(x, sin,
         linewidth=1.5,
         linestyle='-.',
         color='#8E44AD',
         alpha=0.7,
         marker='h')

# cos(x) graphic
plt.plot(x, cos,
         linewidth=1.5,
         linestyle=':',
         color='c',
         alpha=0.7,
         marker='v')

# title of graphic:
plt.title('sin(x) and cos(x) function graphic')
# x-o'q nomi:
plt.xlabel("pi")
# information about lines:
plt.legend(['sin(x)','cos(x)'], loc=0, fontsize='x-large')
plt.grid(color='tomato', alpha=0.5, linestyle='--') # change
plt.show()
```
